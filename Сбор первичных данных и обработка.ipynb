```python
def One_Flat(href):
    data = { }  # Задали пустой словарь, в который мы будем сохранять данные
    data['link'] = href
    # Подгружаем страничку с информацией по квартире
    response = requests.get(href)
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    
    # Вытаскиваем цену на квартиру
    try:
        l = soup.findAll('span', itemprop="price")
        price = l[0].attrs["content"].replace(' ','')
        data['Цена'] = price
    except Exception:
        data['Цена'] = 'NA'
    
    # Вытаскиваем расстояние до метро то ли пешком то ли на машине
    try:
        Do_metro = soup.findAll('span', attrs = {'class':"a10a3f92e9--underground_time--1fKft"})[0].text.strip()
        data['До Метро'] = Do_metro
    except Exception:
        data['До Метро'] = "NA"
    
    
    # Вытаскиваем количество комнат в квартире
    try:
        komnata=soup.findAll('h1', attrs = {'class':"a10a3f92e9--title--2Widg"})[0].text
        data['Комнаты'] = komnata
    except Exception:
        data['Комнаты'] = 'NA'
    # Вытаскиваем район квартиры
    try:
        raion = soup.findAll('a', attrs = {'class':"a10a3f92e9--link--1t8n1 a10a3f92e9--address-item--1clHr"})[2].text
        data['Район'] = raion
    except Exception:
        data['Район'] = 'NA'

    table3=soup.findAll('div', attrs = {'class':"a10a3f92e9--info--3XiXi"})
    if len(table3)>0:
        for row in table3:
            cols3=[row.findAll('div')[1].text,row.findAll('div')[0].text ]
            data[cols3[0]] = cols3[1]
    
    try:
        descr = soup.findAll('p', itemprop="description")[0].text
        data['Описание']=descr
    except Exception:
        data['Описание'] = 'NA'
        
    type_seller=[soup.findAll('div', attrs = {'class':"a10a3f92e9--prop--1F4GC"}),
    soup.findAll('h2', attrs = {'class':"a10a3f92e9--title--BarZi"})] 
    if type_seller[0]==[]:
        if type_seller[1]==[]:
            seller='NA'
        else:
            seller=type_seller[1][0].text
    else:
        seller=type_seller[0][0].text
    data['Тип продавца'] =seller
    
     #Общие характеристики
    table =soup.findAll('section',attrs = {'class':"a10a3f92e9--container--1MHfF"})
    if len(table)>0:
        for row in table:
            cols=[row.findAll('span')[0].text.strip(),row.findAll('span')[1].text.strip() ]
            data[cols[0]] = cols[1]
            
    table1 = soup.findAll('div',attrs = {'class':"a10a3f92e9--item--2Ig2y"})
    if len(table1)>0:
        for row in table1:
            cols2=[row.findAll('div')[0].text,row.findAll('div')[1].text]
            data[cols2[0]] = cols2[1]
    return(data)
```

```python
def Map(part_hrefs):
    # В эту табличку будем собирать данные
    dfr = pd.DataFrame( )
    k = 0 # Это номера наблюдений
    n = len(part_hrefs)
    for item in part_hrefs:
        k = k + 1
        # грузим новое наблюдение/
        dfr1 = pd.DataFrame.from_dict(One_Flat(item),orient='index')
        # присваиваем этому наблюдению номер
        dfr1.columns =[k]
        # закидываем его в итоговую таблицу
        dfr = dfr.join(dfr1, how='outer')
        # Выдавать информацию о том, что сделана каждая десятая итерация! 
        if k%100 == 0:
            print('сделано ', k, ' итераций из', n)
    return(dfr)
```

```python
def Reduce(l):
    df = pd.DataFrame( )  # новый датафрейм, куда мы всё закинем! 
    k = 0
    for item in l:        # Этот цикл уже вам знаком...
        n = item.shape[1] + k
        item.columns = range(k,n)
        df = df.join(item, how = 'outer')
        k = n 
    return(df)  
```

```python
from sklearn.preprocessing import OneHotEncoder as SklearnOneHotEncoder

class OneHotEncoder(SklearnOneHotEncoder):
    def __init__(self, **kwargs):
        super(OneHotEncoder, self).__init__(**kwargs)
        self.fit_flag = False

    def fit(self, X, **kwargs):
        out = super().fit(X)
        self.fit_flag = True
        return out

    def transform(self, X, **kwargs):
        sparse_matrix = super(OneHotEncoder, self).transform(X)
        new_columns = self.get_new_columns(X=X)
        d_out = pd.DataFrame(sparse_matrix.toarray(), columns=new_columns, index=X.index)
        return d_out

    def fit_transform(self, X, **kwargs):
        self.fit(X)
        return self.transform(X)

    def get_new_columns(self, X):
        new_columns = []
        for i, column in enumerate(X.columns):
            j = 0
            while j < len(self.categories_[i]):
                new_columns.append(f'{column}_<{self.categories_[i][j]}>')
                j += 1
        return new_columns
```

```python
df[['Отопление','Газоснабжение','Тип перекрытий','Тип дома','Тип жилья','Тип продавца', 'Парковка']]=df[['Отопление','Газоснабжение','Тип перекрытий','Тип дома','Тип жилья','Тип продавца', 'Парковка']].replace(np.nan, 'Нет инфо')

df['Мусоропровод_dummy'] = np.where(df.Мусоропровод == 'Есть', 1, 0)
df['Лифт_dummy'] = np.where(df.Лифты.notnull(), 1, 0)
df['Авариное'] = np.where(df.Аварийность == 'Да', 1, 0)

df['metro_time']=df['До Метро'].str.extract(r'(\d{1,2})')

df['Кухня']=df['Кухня'].str.extract(r'(\d{1,3})').apply(pd.to_numeric)
df['Жилая']=df['Жилая'].str.extract(r'(\d{1,3})').apply(pd.to_numeric)
df['Общая']=df['Общая'].str.extract(r'(\d{1,3})').apply(pd.to_numeric)
df['Цена']=df['Цена'].str.replace('₽', '').replace(',','.').apply(pd.to_numeric)

def f(a):
    if str(a).find('транспорте')>-1:
        return 'На транспорте близко'
    if str(a).find('пешком')>-1:
        return 'В пешей доступности'
    else:
        return 'Далеко'
    
df['Транспорт']=df['До Метро'].apply(f)

 
#этаж текущий
df['Этаж1']=df['Этаж'].str.extract(r'(\d{1,2}\s)')
df['Этаж1'] = df['Этаж1'].replace(r'\s+','',regex=True)
 
#общее число этаже в доме
df['Этажей']=df['Этаж'].str.extract(r'(\s\d+)')
df['Этажей'] = df['Этажей'].replace(r'\s+','',regex=True)

df[['Этаж1', 'Этажей']] = df[['Этаж1', 'Этажей']].apply(pd.to_numeric)
df['Доля']=df['Этаж1']/df['Этажей']

df['Комната']=df['Комнаты'].str.extract(r'(.)')

df['Собственник'] = np.where(df['Тип продавца'] == 'Собственник', 1, 0)

```
